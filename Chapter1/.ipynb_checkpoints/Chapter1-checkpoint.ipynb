```python
import os
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
DATADIR = '../datasets/'
```

```python
ls -1 ../datasets/
```

```python
df = pd.read_csv(os.path.join(DATADIR, 'whas100.csv'), encoding='shift-jis', index_col=0)
```

```python
sns.set_theme(style='ticks', font='Meiryo', context='paper')
plt.title('追跡時間と年齢の散布図')
sns.scatterplot(data=df, x='追跡日数', y='登録時年齢', hue='生存状態');
```
